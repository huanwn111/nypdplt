```python
#=====计算年龄平均值，过滤空值=====
meanAge = sum(tit['Age'])/len(tit['Age'])#Age列总和/Aqe列个数
print(meanAge)#结果打印非数字NAN，因为里面有空值。
#所以下面要过滤空值
#插播几条isnull函数
isNullAge= pd.isnull(tit['Age'])#Age为空吗？不空的为False，空的为true
print(isNullAge)
print("===只显示Age空的，对象['字段名']['再加一个过滤条件']===")
print(tit['Age'][isNullAge])
#===下面继续，过滤空值再计算===
realAges = tit['Age'][isNullAge==False]#真实有效的年龄字段，过波条件是 空==False，即不空
print("===非空的真实年龄==")
print(realAges)
#realMeanAge = sum(realAges['Age'])/len(realAges['Age'])
realMeanAge = sum(realAges)/len(realAges)#【已经是过滤出来的只有年龄的字段了，就不用再取了，直接计算即可】
print("===平均年龄：====")
print(realMeanAge)
print("==【重点来了！！！：Pandas自带mean()函数，自动过滤空值，算平均值】==")
print(tit['Age'].mean())#【看到没，上面一大堆过滤空，就用一句 对象['字段'].mean()就计算完成了……完全一样，是不是有点怀疑人生，哈哈哈】
```

```python
#===另一实例，计算三个仓的平均价格===
#三个仓 建list遍历，每个等级的价格取出来它的Fare,mean()算平均，每个仓级的结果，存在一个dict里
classes = [1,2,3]
rsDatas = {}
for item in classes:
    datasByClass = tit[tit['Pclass']==item]#遍历仓级，按仓极分开DataFrame
    #【注意上面这里，不是用的if判断，用了过滤条件，当前分类为1时它的Datas取出，再取fare求平均。不好理解就当if理解。
    #这种过滤条件的用法很好，注意使用，在python 不用库里也可以试下】
    avgFareByClass = datasByClass['Fare'].mean()#取票价，求平均
    #追加到结果字典中
    rsDatas[str(item)+'等仓'] = avgFareByClass
    #【注意，给等级拼字符串的时候，要从整形转成str】
print("=====结果=======")    
print(rsDatas)
```

```python
#===函数实例：返回空记录的个数=====
def getNull(obj):
    nullBool = pd.isnull(obj);#判断为空False占位，不空True占位,【返回每位真假bool值】
    nullDatas = obj[nullBool]#【真假占位Bool值传回tit做索引，过波出False空，过波掉True 非空】
    return len(nullDatas)#【统计空记录数，按列统计】
tit.apply(getNull)
```

```python
#===函数应用：取Pclass列 判断tit Pclass 1 2 3 print 仓级===
def showClass(obj):
    classColumn = obj['Pclass'];#筛出Plass列
    #判断返回
    if pd.isnull(classColumn):#为空返回未知，注意这里是圆括号，它是一个函数
        return '未知'
    elif classColumn == 1:
        return '一等仓'
    elif classColumn == 2:
        return '二等仓'
    elif classColumn == 3:
        return '三等仓'
    else:
        return '其它'
    
print(tit.apply(showClass,axis = 1))#【后面这个axis必须有，函数里面用到取列，所以要规定一个轴？】
```

```python
#===函数实例：==
def isChild(obj):
    if obj['Age'] < 18 :
        return True
    else:
        return False
tit.apply(isChild,axis =1)#【看到没，函数里面需传列名的，这里都要加第二参axis，以行统计】
```

```python
import numpy as np

#==函数实例：==
def ageLevel(obj):
    age = obj['Age']
    if pd.isnull(age):
        return '未知'
    elif age < 18:
        return '小孩'
    else:
        return '大人'
    
#===统计大人小孩子的获救率==
print("===按列依次统计====")
print(tit.pivot_table(index = 'Age', values = 'Survived',aggfunc = np.mean))
print("===按小孩大人划分年龄段统计：思路，用上面函数，给样本加一列，标记agelevel，用上面方法统计新加列===")
ageLevelC = tit.apply(ageLevel,axis=1)
print(ageLevelC)
print("===【注意，新加一个新列保存这ageLevelC年龄标记列的方法 tit['新列名']】=ageLevelC==")
tit['ageLevelC'] = ageLevelC;
print("==【开始统计】==")
print(tit.pivot_table(index = 'ageLevelC',values = 'Survived', aggfunc = np.mean))#第一参，关键点：统计的index变了。
```

```python
#==序列的操作，重组新的数据表==
#=把上面电影名和一个评分评台的评分组成一个新表，拿出两个数据看看=
from pandas import Series#需引入Series函数
filmSc = pd.read_csv("fandango_score_comparison.csv")
newFilm = filmSc['FILM'].values
newRt = filmSc['RottenTomatoes'].values
filmSc2 = Series(newRt,index = newFilm)
#两参为上面两列的values(很显示是取属性值重组，还不是序列对象),index=规定以一个列为索引
#newRt不能做索引，可能因为值中有很多重复值，
print(filmSc2)
print("==取0，3索引号样本数据===")
print(filmSc2[[0,3]])
print("==也可直接取film名，达到上面效果，因它设为了index索引===")
print(filmSc2[['Avengers: Age of Ultron (2015)','Do You Believe? (2015)']])
#【注意间隔的，不连续的，要双层[]】
print("==试下索相值为另一列时是否能用电影名取?也可以。一样的==")
```

```python
#==把上面新序列用索引值排序 .index.tolist，sorted，reindex===
#print(filmSc2.sort_values('FILM').reset_index(drop=False))
#上面，用了之前的排序方法sort_values('字段名').reset_index(是否替换还是加新列)
#显然不行，因为没有列名。取到的只是series值。单纯的值有时候也是需要用的数据。
#要达到上面排序效果，
#思路：把【index列取到，tolist,排这个list做新的索引，再传到，对象.reindex(排好序城乡之索引)】
oldIndexList = filmSc2.index.tolist()#【变成list，可用sorted函数排序】
newIndexList = sorted(oldIndexList);
print(filmSc2.reindex(newIndexList))#【redindex('排序列表')相当于按括号里列排序，只不过先排好了再re一下，放进去的意思】
```

```python
#==【好了，重点来了，做完了上面复杂的排序操作，series有自带的排序函数，一步实现上面操作】==
print("===【按值(非index那列)排序 sort_values()】===")
sc1 = filmSc2.sort_values()
print(sc1[0:3])
print("===【按index排序 sort_index()】===")
sc2 = filmSc2.sort_index()
print(sc2[0:3])
```
