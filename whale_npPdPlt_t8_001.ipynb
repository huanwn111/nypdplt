```python
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
import pandas as pd
```

```python
data = np.arange(10)
data
```

```python
plt.plot(data)
```

```python
#画分成组图
#建组
fig = plt.figure()
#生几三个子画布
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
#画图
plt.plot(np.random.randn(50).cumsum(),'k--')#默认从最后一张画布上画4
#在上面ax1子画布画直方图hist
_ = ax1.hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
#在ax2上画散点图scatter
ax2.scatter(np.arange(30),np.arange(30) + 3*np.random.randn(30))
```

```python
#画图
plt.plot(np.random.randn(50).cumsum(),'k--')
```

```python
#用plt.subplots （它创建figure并包含subplot对象的numpy数组） 改写上面
fig,axes = plt.subplots(2,3)
axes
_ = axes[0,1].hist(np.random.randn(100),bins=20,color='k',alpha=0.3)
axes[1,2].scatter(np.arange(30),np.arange(30) + 1*np.random.randn(30))
```
